```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
   
        
    num_products_view_2 = 0
    scroll_range = 1700
    driver.execute_script('window.scrollTo(0, ' + str(scroll_range) + ')')
    sleep(5)
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(100):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "vtex-product-summary-2-x-nameContainer flex items-center justify-center pv6"})))
        print(num_products_view_1)
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                the_name_id = soup.find_all(attrs={"class": "vtex-search-result-3-x-galleryItem vtex-search-result-3-x-galleryItem--normal pa4"})[counter].find_all(attrs={"class": "vtex-product-summary-2-x-productBrand vtex-product-summary-2-x-brandName t-body"})[0].text
                try: 
                    the_price_id = soup.find_all(attrs={"class": "vtex-search-result-3-x-galleryItem vtex-search-result-3-x-galleryItem--normal pa4"})[counter].find_all(True, {"class": "vtex-productShowCaseListPrice"})[0].text
                except: 
                    try:
                        the_price_id = soup.find_all(attrs={"class": "vtex-search-result-3-x-galleryItem vtex-search-result-3-x-galleryItem--normal pa4"})[counter].find_all(True, {"class": "vtex-productShowCasePrice"})[0].text
                    except:
                        the_price_id = "no price available"
                print(the_price_id)
                the_brand_id = the_name_id.split()
                the_brand_id = the_brand_id[0] + " " + the_brand_id[1]

                the_volume_id = the_name_id.split()
                for i in range(0,len(the_volume_id)):
                    if "sulas" in the_volume_id[i]:
                        the_volume = the_volume_id[i-1] + " " + the_volume_id[i]
                    elif 'dades' in the_volume_id[i]: 
                        the_volume = the_volume_id[i-1] + " " + the_volume_id[i]
                    elif ('g' in the_volume_id[i] or 'kg' in the_volume_id[i]) and any(ch.isdigit() for ch in the_volume_id[i]):
                        the_volume = the_volume_id[i]
                    elif ('ml' in the_volume_id[i] or 'l' in the_volume_id[i]) and any(ch.isdigit() for ch in the_volume_id[i]):
                        the_volume = the_volume_id[i]
                    else: 
                        the_volume = "no volume available"
                print(the_volume)
                filename = "Bompreco"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per product into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume,the_name_id, the_price_id])
                
                scroll_range += 100
                driver.execute_script('window.scrollTo(0, ' + str(scroll_range) + ')')
        
        
            
        #break als er geen "next" button gevonden kan worden  
        sleep(3)
        new_element = driver.find_element(By.CSS_SELECTOR, 'body > div.render-container.render-route-store-search > div > div.vtex-store__template.bg-base > div > div:nth-child(6) > div > div > section > div.relative.justify-center.flex > div > div.vtex-flex-layout-0-x-flexRow > div > div.pr0.items-stretch.flex-grow-1.flex > div > div:nth-child(7) > div > div > div > div > div > button')
        driver.execute_script("arguments[0].click();", new_element)
        sleep(1)
            
       #update total number of products in current view to compare with number of products in view before scrolling
        num_products_view_2 = int(len(soup.find_all(attrs={"class": "vtex-search-result-3-x-galleryItem vtex-search-result-3-x-galleryItem--normal pa4"})))
        print(num_products_view_2)
        
        
        
scroll_page("Bompreco")
```
