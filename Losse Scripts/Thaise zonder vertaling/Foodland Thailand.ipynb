```python
def scroll_page(retailer):    
    
    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    scroll_range = 0
    num_products_view_2 = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(200):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "product_content_inner"})))
        print(num_products_view_1)
        
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                the_name_id = soup.find_all(attrs={"class": "product_content_inner"})[counter].find_all(attrs={"class": "product_name"})[0].text

                
                the_price_id = soup.find_all(attrs={"class": "product_content_inner"})[counter].find_all(attrs={"class": "price_box"})[0].text
                the_price_id = the_price_id.replace("\n", "")
                the_price_id = the_price_id.replace("  ", "")
                the_price_id = the_price_id.split()
                try: 
                    the_price_id = the_price_id[1]
                except:
                    the_price_id = the_price_id[0]  
                print(the_price_id)
                
                
                the_brand_id = the_name_id.split()
                try: 
                    the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
                except:
                    the_brand_id = the_brand_id[0]
                
                
                the_volume_id = soup.find_all(attrs={"class": "product_content_inner"})[counter].find_all(attrs={"class": "memberpoint"})[0].text
                the_volume_id = the_volume_id.replace("\n", "")
                the_volume_id = the_volume_id.replace("  ", "")
                print(the_volume_id)
                
                filename = "Foodland Thailand"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per product into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-16') as csv_file:
                    writer = csv.writer(csv_file, delimiter = "\t")
                    writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
            
                scroll_range += 100
                driver.execute_script('window.scrollTo(0, ' + str(scroll_range) + ')')
                
        #break als er geen "next" button gevonden kan worden  
        sleep(0.5)
        
        #update total number of products in current view to compare with number of products in view before scrolling
        num_products_view_2 = int(len(soup.find_all(attrs={"class": "product_content_inner"})))

scroll_page("Foodland Thailand")
```
