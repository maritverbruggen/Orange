```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())

from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from datetime import datetime
from selenium.webdriver.common.keys import Keys

import pandas as pd
import statistics
import csv
from time import sleep
import math
```

```python
driver.get('https://www.allonline.7eleven.co.th/beverage-instant-drink/coffee/')
driver.maximize_window()
sleep(3)
```

```python
def scroll_page(retailer):    
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    num_products_view_2 = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(20):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "item-list-wrapper"})))
        print(num_products_view_1)
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id 
        for counter in range(num_products_view_2, num_products_view_1):
            the_name_id = soup.find_all(attrs={"class": "item-list-wrapper"})[counter].find_all(attrs={"class": "item description item-description-cls-mobile"})[0].text
            the_name_id = the_name_id.replace("\n", "")
            the_name_id = the_name_id.replace("  ", "")
            print(the_name_id)
            try: 
            
                the_price_id = soup.find_all(attrs={"class": "item-list-wrapper"})[counter].find_all(attrs={"class": "price-bottom"})[0].text
                the_price_id = the_price_id.split()
                the_price_id = the_price_id[2] + the_price_id[3]

            except: 
                the_price_id = soup.find_all(attrs={"class": "item-list-wrapper"})[counter].find_all(attrs={"class": "price-bottom"})[0].text
                the_price_id = the_price_id.split()
                the_price_id = the_price_id[0] + the_price_id[1]
            print(the_price_id)   
            
            the_brand_id = the_name_id.split()
            the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
            
            the_volume_id = the_name_id.split()
            for i in range(0,len(the_volume_id)):
                if any(ch.isdigit() for ch in the_volume_id[i])== True:
                    the_volume = the_volume_id[i:]
                    the_volume = ' '.join([str(elem) for elem in the_volume])
                    break
                else: 
                    the_volume = "no volume found"
            print(the_volume)
            
            filename = "7Eleven Thailand"
            fullpath = "../Output/" + str(filename) + ".csv"
            #write variables per product into csv file 
            with open(fullpath, mode='a', newline='', encoding = "utf-16") as csv_file:
                writer = csv.writer(csv_file, delimiter = "\t")
                writer.writerow([retailer, the_brand_id, the_volume,the_name_id, the_price_id])
            
        #break als er geen "next" button gevonden kan worden  
        sleep(5)
        try:
            new_element = driver.find_element(By.CSS_SELECTOR, '#page > div.container > div.kes.home > div.relative > div > div.col-xs-12.col-sm-9 > div > div:nth-child(11) > div > div > div > div > div > div:nth-child(9) > div.col-md-8.col-xs-12.hidden-print > ul > li.next > span')
            new_element.click()
            sleep(1)
    
        except: 
            print('kan niet vinden')
            break
                
scroll_page("7Eleven Thailand")
```
