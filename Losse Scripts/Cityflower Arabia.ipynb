```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
  
    num_products_view_2 = 0
    scroll_range = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(150):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of wines in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "jss7 MuiBox-root css-1g9jsrx"})))
        print(num_products_view_1)
        
        #create attributes the_name_id, the_price_id, the_reviews_id and the_rating id
        for counter in range(num_products_view_2, num_products_view_1):
            the_name_id = soup.find_all(attrs={"class": "jss7 MuiBox-root css-1g9jsrx"})[counter].find_all(attrs={"class": "jss11"})[0].text
            the_name_id = the_name_id.replace("\n", "")
            the_name_id = the_name_id.replace("  ", "")
            print(the_name_id)
            
            the_price_id = soup.find_all(attrs={"class": "jss7 MuiBox-root css-1g9jsrx"})[counter].find_all(attrs={"class": "jss13"})[0].text
            the_price_id = the_price_id.replace("\n", "")
            the_price_id = the_price_id.replace("  ", "")
            print(the_price_id)
            
            the_brand_id = the_name_id.split()
            the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
            
            the_volume_id = the_name_id.split()
            for i in range(0,len(the_volume_id)):
                if any(ch.isdigit() for ch in the_volume_id[i])== True:
                    the_volume = the_volume_id[i:]
                    the_volume = ' '.join([str(elem) for elem in the_volume])
                    break
                else: 
                    the_volume = "no volume found"
            print(the_volume)                                        

            filename = "Cityflower Arabia"
            fullpath = "../Output/" + str(filename) + ".csv"
            #write variables per wine into csv file 
            with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                writer = csv.writer(csv_file)
                writer.writerow([retailer, the_brand_id, the_volume,the_name_id, the_price_id])
                
            scroll_range += 200
            driver.execute_script('window.scrollTo(0, ' + str(scroll_range) + ')')
            
            num_products_view_2 = int(len(soup.find_all(attrs={"class": "jss7 MuiBox-root css-1g9jsrx"})))
scroll_page("Cityflower Arabia")
```
