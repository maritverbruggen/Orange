```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    lengte = 0
    num_products_view_2 = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(20):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of wines in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "product-card"})))
        print(num_products_view_1)
        #create attributes the_name_id, the_price_id, the_reviews_id and the_rating id
        for counter in range(num_products_view_2, num_products_view_1):
            the_name_id = soup.find_all(attrs={"class": "product-card"})[counter].find_all(attrs={"class": "product-card__name"})[0].text
            the_name_id = the_name_id.replace("\n", "")
            the_name_id = the_name_id.replace("  ", "")            
            print(the_name_id)
            try: 
                the_price_id = soup.find_all(attrs={"class": "product-card"})[counter].find_all(attrs={"class": "product-card__price--old loyalty-points__price--default"})[0].text
                the_price_id = the_price_id.replace("\n", "")
                the_price_id = the_price_id.replace("  ", "")
            except: 
                the_price_id = soup.find_all(attrs={"class": "product-card"})[counter].find_all(attrs={"class": "product-card__price--current loyalty-points__price--default"})[0].text
                the_price_id = the_price_id.replace("\n", "")
                the_price_id = the_price_id.replace("  ", "")
            print(the_price_id)
            the_brand_id = the_name_id.split()
            the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
            the_volume_id = the_name_id.split()
            if "Kapseln" in the_volume_id:
                i = int(the_volume_id.index("Kapseln"))
                the_volume_id = the_volume_id[i-1] + " " + the_volume_id[i] 
            else: 
                the_volume_id = soup.find_all(attrs={"class": "product-card"})[counter].find_all(attrs={"class": "product-card__weight"})[0].text
            the_volume_id = the_volume_id.replace("\n", "")
            the_volume_id = the_volume_id.replace("  ", "")
            print(the_volume_id)
            filename = "Combi"
            fullpath = "../Output/" + str(filename) + ".csv"
            #write variables per wine into csv file 
            with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                writer = csv.writer(csv_file)
                writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])  
            lengte += 1 
        #break als er geen "next" button gevonden kan worden  
        sleep(5)
    
        new_element = driver.find_element(By.CSS_SELECTOR, '#product-list-210001147 > header.nav-category.nav-category--top > div.nav-category__item.nav-category__item__pagination > ul > li:nth-child(3) > a > i')
        new_element.click()
        sleep(1)

 
        if lengte >= total_results: 
            break 
        
scroll_page("Combi")
```
