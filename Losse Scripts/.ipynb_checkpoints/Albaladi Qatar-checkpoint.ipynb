```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())

from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from datetime import datetime
from selenium.webdriver.common.keys import Keys

import pandas as pd
import statistics
import csv
from time import sleep
import math
```

```python
from bs4 import BeautifulSoup
res = driver.page_source.encode('utf-8')
soup = BeautifulSoup(res, "html.parser")
```

```python
driver.get('https://albaladi-qatar.com/index.php?route=product/category&path=394_404')
driver.maximize_window()
```

```python
amount_of_clicks = driver.find_element(By.CSS_SELECTOR, '#content > div.custom-category > div.tool-bar-bottom > div > div.col-sm-6.text-right').text
amount_of_clicks = amount_of_clicks.replace("(", "")
amount_of_clicks = amount_of_clicks.split()
amount_of_clicks = amount_of_clicks[-2]
amount_of_clicks = int(amount_of_clicks)
print(amount_of_clicks)
```

```python
def scroll_page(retailer):    
    
        
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    num_products_view_2 = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(amount_of_clicks):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "product-thumb"})))
        
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                
                the_name_id = soup.find_all(attrs={"class": "product-thumb"})[counter].find_all(attrs={"class": "caption"})[0].find_all("h4")[0].text
                the_name_id = the_name_id.replace("\n", "")
                the_name_id = the_name_id.replace("  ", "")
                print(the_name_id)
               
                the_price_id = soup.find_all(attrs={"class": "product-thumb"})[counter].find_all(attrs={"class": "price"})[0].text
                the_price_id = the_price_id.replace("\n", "")
                the_price_id = the_price_id.replace("  ", "")
                the_price_id = the_price_id.replace("\t", "")
                print(the_price_id)
                try: 
                    
                    the_brand_id = soup.find_all(attrs={"class": "product-thumb"})[counter].find_all(attrs={"class": "manufacture-product"})[0].text
                    the_brand_id = the_brand_id.replace("\n", "")
                    the_brand_id = the_brand_id.replace("  ", "")
                    print(the_brand_id)
                except:
                    the_brand_id = the_name_id.split()
                    the_brand_id = the_brand_id[0] 
                
                the_volume_id =  the_name_id.split()
                the_volume_id = the_volume_id[-2] + " " + the_volume_id[-1]
                print(the_volume_id)
                filename = "Albaladi Qatar"
                fullpath = "../Output/" + str(filename) + ".csv"
                
                #write variables per product into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
            
        #break als er geen "next" button gevonden kan worden  
        sleep(3)
        try:
            new_element = driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[3]/div/div[1]/ul/li[12]/a')
            new_element.click()
            sleep(2)
        except:
      
            new_element = driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[3]/div/div[1]/ul/li[10]/a')
            new_element.click()
            sleep(2)

scroll_page("Albaladi Qatar")
```
