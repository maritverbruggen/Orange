```python
#import relevant packages used in this script

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
driver = webdriver.Chrome(ChromeDriverManager().install())


from selenium.webdriver import ActionChains
action = ActionChains(driver)
from selenium.webdriver.common.by import By
from datetime import datetime

import pandas as pd
import statistics
import csv
from time import sleep
import math
```

```python
from bs4 import BeautifulSoup
res = driver.page_source.encode('utf-8')
soup = BeautifulSoup(res, "html.parser")
```

```python
driver.get("https://safarihypermarket.ae/products/?sub=1203&category=1959&parent=1004")
driver.maximize_window()
element = driver.find_element(By.CSS_SELECTOR,'body > section.landing-page > div > div > div > div.country-listing-wrap > ul > li:nth-child(1) > a')
element.click()
sleep(3)
element = driver.find_element(By.CSS_SELECTOR, 'body > section.landing-page > div > div > div > div.changeLocation-modal > ul > li:nth-child(1) > a')
element.click()
sleep(3)
inputElement = driver.find_element(By.CSS_SELECTOR, "#search_data1")
inputElement.send_keys('qatar')
inputElement.send_keys(Keys.ENTER)
sleep(3)
element = driver.find_element(By.CSS_SELECTOR, '#autoSuggestionsList12 > li:nth-child(1) > a > span.zone-name')
element.click()
sleep(3)
element = driver.find_element(By.CSS_SELECTOR, '#navbarText > ul > li:nth-child(4) > a')
action.move_to_element(element).perform()
sleep(3)
element = driver.find_element(By.CSS_SELECTOR, '#mCSB_15_container > li:nth-child(6) > a')
element.click()
sleep(3)
```

```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    num_products_view_2 = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(30):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of wines in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "product"})))
        print(num_products_view_1)
        
        #create attributes the_name_id, the_price_id, the_reviews_id and the_rating id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                
                the_name_id = soup.find_all(attrs={"class": "product"})[counter].find_all("h5")[0].find_all("a")[0].text
                the_name_id = the_name_id.replace("\n", "")
                the_name_id = the_name_id.replace("  ", "")
                print(the_name_id)
            
                the_price_id = soup.find_all(attrs={"class": "product"})[counter].find_all(attrs={"class": "product-footer"})[0].find_all("p")[0].text
                the_price_id = the_price_id.replace("\n", "")
                the_price_id = the_price_id.replace("  ", "")
                print(the_price_id)
                
                the_brand_id = the_name_id.split()
                the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
                print(the_brand_id)
                
                the_volume_id = the_name_id.split()
                for i in range(0,len(the_volume_id)):
                    
                    if "X" == the_volume_id[i]:
                        the_volume = the_volume_id[i-1] + " " + the_volume_id[i] + " " + the_volume_id[i+1]
                        break
                    elif "`S" in the_volume_id[i] and any(ch.isdigit() for ch in the_volume_id[i]):
                        the_volume = the_volume_id[i]
                        break 
                    elif "x" in the_volume_id[i] and any(ch.isdigit() for ch in the_volume_id[i]):
                        the_volume = the_volume_id[i]
                        break
                    elif ('gm' in the_volume_id[i] or 'g' in the_volume_id[i] or 'Kg' in the_volume_id[i]) and any(ch.isdigit() for ch in the_volume_id[i]):
                        the_volume = the_volume_id[i]
                        break
                    else: 
                        the_volume = the_volume_id[-2] + " " + the_volume_id[-1]
                
                filename = "Safari Qatar"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per wine into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume,the_name_id, the_price_id])
            
        #break als er geen "next" button gevonden kan worden  
        sleep(3)
        try:
            new_element = driver.find_element(By.CSS_SELECTOR, '#new > div.col-md-12 > div > div > a.btn.btn-secondary.mb-3.loadmore.proc1')
            new_element.click()
            sleep(1)

        except:
            print('kan niet vinden')
            break
        
        #update total number of products in current view to compare with number of products in view before scrolling
        num_products_view_2 = int(len(soup.find_all(attrs={"class": "product"})))
        print(num_products_view_2)
```

```python
scroll_page("Safari Qatar")
```
