```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    
    #create scraping_round variable that counts the amount of times all the wines are being scraped 
    scraping_round = 1
        
    num_products_view_2 = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(20):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of wines in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "wrapper"})))
        
        #create attributes the_name_id, the_price_id, the_reviews_id and the_rating id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                the_name_id = soup.find_all(attrs={"class": "wrapper"})[counter].find_all(attrs={"class": "product-title at-all-productName-lbl"})[0].text
                the_price_id = soup.find_all(attrs={"class": "wrapper"})[counter].find_all(attrs={"class": "price at-product-price_lbl"})[0].text
                the_brand_id = the_name_id.split()
                the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
                the_volume_id = the_name_id.split()
                if any(ch.isdigit() for ch in the_volume_id[-2]) == True: 
                    the_volume_id = the_volume_id[-2] + the_volume_id[-1]
                elif any(ch.isdigit() for ch in the_volume_id[-1]) == True: 
                    the_volume_id = the_volume_id[-1]
                else: 
                    the_volume_id = "No volume available"
                print(the_volume_id)
                filename = "Aldi Duitsland"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per wine into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
                       
            scraping_round += 1
            
        #break als er geen "next" button gevonden kan worden  
        sleep(3)
        try:
            new_element = driver.find_element(By.XPATH, '//*[@id="showMore"]')
            new_element.click()
            sleep(1)

        except:
            print('kan niet vinden')
            break
        
        #update total number of wines in current view to compare with number of products in view before scrolling
        num_products_view_2 = int(len(soup.find_all(attrs={"class": "wrapper"})))
        print(num_products_view_2)

    
scroll_page("Aldi Duitsland")
```

```python
def scroll_page(retailer):    
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    

        
    num_products_view_2 = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(100):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "product__data"})))
        print(num_products_view_1)
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                the_name_id = soup.find_all(attrs={"class": "product__data"})[counter].find_all(attrs={"class": "product__title"})[0].text
                the_name_id = the_name_id.replace("\n", "")
                the_name_id = the_name_id.replace("  ", "")
                print(the_name_id)
                try:
                    the_price_id = soup.find_all(attrs={"class": "product__data"})[counter].find_all(attrs={"class": "price"})[0].text
                    the_price_id = the_price_id.replace("\n", "")
                    the_price_id = the_price_id.replace("  ", "")
                except: 
                    the_price_id = "no price available"
                print(the_price_id)
                the_brand_id = the_name_id.split()
                try: 
                    the_brand_id = the_brand_id[0] + " " + the_brand_id[1] + " " + the_brand_id[2]
                except: 
                    the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
                the_volume_id = the_name_id.split()
                for i in range(0,len(the_volume_id)):
                    if any(ch.isdigit() for ch in the_volume_id[i])== True:
                        the_volume = the_volume_id[i:]
                        the_volume = ' '.join([str(elem) for elem in the_volume])
                        break
                    else: 
                        the_volume = "no volume found"
                print(the_volume)
                filename = "Kaufland"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per product into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume,the_name_id, the_price_id])
                       
            scraping_round += 1
            
        #break als er geen "next" button gevonden kan worden  
        sleep(3)
        try:
            new_element = driver.find_element(By.CSS_SELECTOR, '#page > div > div.main-content__central-container > div > div.main-content__results > nav > button:nth-child(11) > span > svg')
            new_element.click()
            sleep(1)

        except:
            print('kan niet vinden')
            break
    
scroll_page("Kaufland")
```
