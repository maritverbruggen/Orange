```python
#import relevant packages used in this script

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())

from selenium.webdriver import ActionChains
# create action chain object
action = ActionChains(driver)
from selenium.webdriver.common.by import By
from datetime import datetime

import pandas as pd
import statistics
import csv
from time import sleep
import math
```

```python
driver.get("https://www.ah.nl/producten/frisdrank-sappen-koffie-thee/koffie")
driver.maximize_window()
element = driver.find_element(By.XPATH,'//*[@id="accept-cookies"]')
element.click()
sleep(5)
```

```python
#find total amount of times to click "load more"
total_results = driver.find_element(By.XPATH,'//*[@id="start-of-content"]/div[3]/span').text
splitted_results = total_results.split()
total_products = int(splitted_results[3])
products_in_window = int(splitted_results[0])
amount_of_clicks = int(splitted_results[3])/int(splitted_results[0])
amount_of_clicks = math.ceil(amount_of_clicks)
print(total_products)
print(products_in_window)
print(amount_of_clicks)
```

```python
def scroll_page(retailer):    
    

    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
        
    num_products_view_2 = 0
    number = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(amount_of_clicks):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "product-card-portrait_root__sZL4I product-grid-lane_gridItem__eqh9g"})))
        
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                the_name_id = soup.find_all(attrs={"class": "product-card-portrait_root__sZL4I product-grid-lane_gridItem__eqh9g"})[counter].find_all(attrs={"class": "title_root__2b4w2 product-card-portrait_title__14Jej"})[0].text
                the_price_id = soup.find_all(attrs={"class": "product-card-portrait_root__sZL4I product-grid-lane_gridItem__eqh9g"})[counter].find_all(True, {"class": ["price-amount_root__37xv2 price-amount_highlight__3WjBM price_amount__2Gk9i price_highlight__3B97G", "price-amount_root__37xv2 price-amount_was__1PrUY price_amount__2Gk9i price_was__1PJTn"]})[0].text
                the_brand_id = the_name_id.split()
                try: 
                    the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
                except:
                    try:
                        the_brand_id = the_brand_id[0]
                    except:
                        the_brand_id = "no brand available"
                        
                the_volume_id = soup.find_all(attrs={"class": "product-card-portrait_root__sZL4I product-grid-lane_gridItem__eqh9g"})[counter].find_all(True, {"class": "price_unitSize__8gRVX"})[0].text           
            
                filename = "Albert Heijn"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per product into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
                number+=1   
                
            print(number)
        #break als er geen "next" button gevonden kan worden  
        sleep(3)
        try:
            new_element = driver.find_element(By.XPATH, '//*[@id="start-of-content"]/div[3]/button/span')
            new_element.click()
            sleep(1)

        except:
            print('kan niet vinden')
            break
        
        #update total number of products in current view to compare with number of products in view before scrolling
        num_products_view_2 = int(len(soup.find_all(attrs={"class": "product-card-portrait_root__sZL4I product-grid-lane_gridItem__eqh9g"})))
        print(num_products_view_2)

        #break loop if all products are scraped
        if num_products_view_2 >= total_products: 
            break
scroll_page("Albert Heijn")
```

```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    
    num_products_view_2 = 0
    page = 0 
    scroll_range = 0 
    
    for _ in range(amount_of_clicks):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of wines in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "c-product-tile"})))
        
        #create attributes the_name_id, the_price_id, the_reviews_id and the_rating id
        for counter in range(num_products_view_2, num_products_view_1):
            the_name_id = soup.find_all(attrs={"class": "c-product-tile"})[counter].find_all(attrs={"class": "c-product-tile__meta"})[0].text
            the_name_id = the_name_id.replace("\n", "")
            the_name_id = the_name_id.replace("  ", "")
            print(the_name_id)
            
            try: 
                         
                the_price_id = soup.find_all(attrs={"class": "c-product-tile"})[counter].find_all(attrs={"class": "c-price"})[0].text
                the_price_id = the_price_id.replace("\n", "")
                the_price_id = the_price_id.replace("  ", "")
                print(the_price_id)
            except:
                the_price_id = "no price available"
                
            the_brand_id = the_name_id.split()
            the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
            print(the_brand_id)
            
            the_volume = soup.find_all(attrs={"class": "c-product-tile"})[counter].find_all(attrs={"class": "c-product-tile__meta"})[0].text
            
            filename = "Spar Nederland"
            fullpath = "../Output/" + str(filename) + ".csv"
            #write variables per wine into csv file 
            with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                writer = csv.writer(csv_file)
                writer.writerow([retailer, the_brand_id, the_volume,the_name_id, the_price_id])
                
            scroll_range += 160
            driver.execute_script('window.scrollTo(0, ' + str(scroll_range) + ')')
            sleep(0.2)
        
        page+=1 
        child = page + 1 
          
        try: 
            new_element = driver.find_element(By.CLASS_NAME,'is-nexy')
            action.move_to_element(new_element).click().perform()
            
            sleep(1)
        except:
            print("kan niet vinden")
            break 
             

        
scroll_page("Spar Nederland")
```
