```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    
    #create scraping_round variable that counts the amount of times all the wines are being scraped 
    scraping_round = 1
    lengte = 0
    num_products_view_2 = 0
    number = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(20):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of wines in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "shelfProductTile-information"})))
        print(num_products_view_1)
        #create attributes the_name_id, the_price_id, the_reviews_id and the_rating id
        for counter in range(num_products_view_2, num_products_view_1):
            the_name_id = soup.find_all(attrs={"class": "shelfProductTile-information"})[counter].find_all(attrs={"class": "shelfProductTile-description"})[0].text
            try: 
                the_price_id = soup.find_all(attrs={"class": "shelfProductTile-information"})[counter].find_all(attrs={"class": "price-was price--large ng-star-inserted"})[0].text
            except:
                try:
                    the_price_id = soup.find_all(attrs={"class": "shelfProductTile-information"})[counter].find_all(attrs={"class": "price price--large ng-star-inserted"})[0].text
                except: 
                    the_price_id = "Out of Stock, No Price Available"
            print(the_name_id)
            the_brand_id = the_name_id.split()
            the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
            the_volume_id = the_name_id.split()
            for i in range(0,len(the_volume_id)):
                if 'ack' in the_volume_id[i] and any(ch.isdigit() for ch in the_volume_id[i-1]):
                    the_volume = the_volume_id[i-1] + " " + "Pack"
                elif 'g' in the_volume_id[i] or 'kg' in the_volume_id[i]:
                    the_volume = the_volume_id[i]
                else: 
                    next 
            for i in range(0,len(the_volume_id)-1): 
                if 'ack' in the_volume_id[i] and any(ch.isdigit() for ch in the_volume_id[i+1]):
                        the_volume = the_volume_id[i+1] + " " + "Pack"
                else: 
                    next 
            print(the_volume)
                
            filename = "Woolworths"
            fullpath = "../Output/" + str(filename) + ".csv"
        
            with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                writer = csv.writer(csv_file)
                writer.writerow([retailer, the_brand_id, the_volume,the_name_id, the_price_id])

            number +=1
        #break als er geen "next" button gevonden kan worden  
        sleep(3)
        try:
            new_element = driver.find_element(By.CSS_SELECTOR, '#search-content > div > shared-paging > div > div.paging-section > a.paging-next.ng-star-inserted > span.next-marker')
            new_element.click()
            sleep(25)

        except:
            print('kan niet vinden')
            print(number)
            break
            
scroll_page("Woolworths")
```
