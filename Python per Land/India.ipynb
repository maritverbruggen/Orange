```python
sleep(10)

total_results = driver.find_element(By.XPATH,'//*[@id="deck"]/div[2]/div/h2').text
print(total_results)
R = ["(", ")"]
total_results = list(total_results)
N = [i for i in range(len(total_results)) if total_results[i] in R]
totalresults = []
for i in range(N[0]+1, N[1]):
    totalresults += total_results[i]
results = ""
results = results.join(totalresults)
results = int(results)
print(results)
```

```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    
    #create scraping_round variable that counts the amount of times all the wines are being scraped 
    scraping_round = 1
        
    num_products_view_2 = 0
    scroll = 500
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(20):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of wines in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "item prod-deck row ng-scope"})))
        print(num_products_view_1)
        
        #create attributes the_name_id, the_price_id, the_reviews_id and the_rating id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                the_name_id = soup.find_all(attrs={"class": "item prod-deck row ng-scope"})[counter].find_all(attrs={"ng-bind": "vm.selectedProduct.p_desc"})[0].text
                print(the_name_id)
                try: 
                    the_price_id = soup.find_all(attrs={"class": "item prod-deck row ng-scope"})[counter].find_all(attrs={"class": "mp-price ng-scope"})[0].text
                except: 
                    the_price_id = soup.find_all(attrs={"class": "item prod-deck row ng-scope"})[counter].find_all(attrs={"class": "discnt-price"})[0].text
                print(the_price_id)
                the_brand_id = soup.find_all(attrs={"class": "item prod-deck row ng-scope"})[counter].find_all(attrs={"ng-bind":"vm.selectedProduct.p_brand"})[0].text
                print(the_brand_id)
                the_volume_id = soup.find_all(attrs={"class": "item prod-deck row ng-scope"})[counter].find_all(attrs={"ng-bind": "vm.selectedProduct.w"})[0].text           
                print(the_volume_id)
                filename = "BigBasket"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per wine into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
                       
            scraping_round += 1
        #break als er geen "next" button gevonden kan worden  
        sleep(3)
        driver.execute_script("window.scrollTo(0, 7000);")
        sleep(10)
        try:
            new_element = driver.find_element(By.CSS_SELECTOR, '#dynamicDirective > product-deck > section > div.col-md-9.wid-fix.clearfix.pl-wrap > div.col-xs-12.product-deck-container.pad-0 > div.show-more > button')
            new_element.click()
            sleep(1)

        except:
            print('kan niet vinden')
            
        
        #update total number of wines in current view to compare with number of products in view before scrolling
        num_products_view_2 = int(len(soup.find_all(attrs={"class": "item prod-deck row ng-scope"})))
        print(num_products_view_2)

        #break loop if all products are scraped
        if num_products_view_2 >= results: 
            break
scroll_page("BigBasket")
```
