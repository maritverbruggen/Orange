```python
driver.get('https://www.tops.co.th/en/beverages/coffee')
driver.maximize_window()
sleep(10)
```

```python
def scroll_page(retailer):    
    
  
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    num_products_view_2 = 0
  
    for _ in range(100):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "product-item"})))
        print(num_products_view_1)
        
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id 
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                the_name_id = soup.find_all(attrs={"class": "product-item"})[counter].find_all(attrs={"class": "product-item--title"})[0].text
                print(the_name_id)
        
                the_price_id = soup.find_all(attrs={"class": "product-item"})[counter].find_all(attrs={"class": "mt-price-number"})[0].text
             
                the_brand_id = the_name_id.split()
                the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
                
                the_volume_id = the_name_id.split()
                for i in range(0,len(the_volume_id)):
                    if any(ch.isdigit() for ch in the_volume_id[i])== True:
                        the_volume = the_volume_id[i]
                        break
                    else: 
                        the_volume = "no volume available"
              
            
                filename = "Tops_Thailand"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per product into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume,the_name_id, the_price_id])
            
                scroll_range += 500
                driver.execute_script('window.scrollTo(0, ' + str(scroll_range) + ')')
                sleep(0.2)
                
        #update total number of products in current view to compare with number of products in view before scrolling
        num_products_view_2 = int(len(soup.find_all(attrs={"class": "product-item"})))
        print(num_products_view_2)

        
scroll_page("Tops_Thailand")
```
