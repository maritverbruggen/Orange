```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    num_products_view_2 = 0
    lengte = 0 
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(20):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20"})))
        print(num_products_view_1)
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id
        for counter in range(num_products_view_2, num_products_view_1):
            the_name_id = soup.find_all(attrs={"class": "sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20"})[counter].find_all(attrs={"class": "a-size-base-plus a-color-base a-text-normal"})[0].text
            print(the_name_id)
            try: 
                the_price_id = soup.find_all(attrs={"class": "sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20"})[counter].find_all(attrs={"class": "a-offscreen"})[0].text
            except: 
                the_price_id = "no price available"
            the_brand_id = the_name_id.split()
            the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
            
            the_volume_id = the_name_id.split()
            for i in range(0,len(the_volume_id)):
                if any(ch.isdigit() for ch in the_volume_id[i])== True:
                    the_volume = the_volume_id[i:]
                    the_volume = ' '.join([str(elem) for elem in the_volume])
                    break
                else: 
                    the_volume = "no volume found"
            
            filename = "Amazon"
            fullpath = "../Output/" + str(filename) + ".csv"
            #write variables per product into csv file 
            with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                writer = csv.writer(csv_file)
                writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
            lengte +=1
        print(lengte)
            
        #break als er geen "next" button gevonden kan worden  
        sleep(5)
        try:
            new_element = driver.find_element(By.LINK_TEXT, 'Next')
            new_element.click()
            sleep(1)

        except:
            print('kan niet vinden')
            break
scroll_page("Amazon")
```
