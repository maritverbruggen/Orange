```python
def scroll_page(retailer):    
    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    
    num_products_view_2 = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(20):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "sku-item"})))
        
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id
        for counter in range(num_products_view_2, num_products_view_1):
            the_name_id = soup.find_all(attrs={"class": "sku-item"})[counter].find_all(attrs={"class": "p-name"})[0].text
            the_name_id = the_name_id.replace("\n", "")

            the_price_id = soup.find_all(attrs={"class": "sku-item"})[counter].find_all(attrs={"class": "p-price-jd"})[0].text
            if the_price_id == "":
                the_price_id = soup.find_all(attrs={"class": "sku-item"})[counter].find_all(attrs={"class": "p-price-zj"})[0].text
            the_price_id = the_price_id.replace("\n", "")
            the_price_id = the_price_id.replace("  ", "")
            the_brand_id = the_name_id.split()
            try: 
                the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
            except:
                the_brand_id = the_name_id
            the_volume_id = the_name_id
            try:
                i = [x.isdigit() for x in the_volume_id].index(True)
                the_volume_id = the_volume_id[i:] 
            except:
                the_volume_id = "No Volume Available"
            print(the_volume_id)
            filename = "JD.JD"
            fullpath = "../Output/" + str(filename) + ".csv"
            #write variables per product into csv file 
            with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                writer = csv.writer(csv_file)
                writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
            
        sleep(10)
        try:
            new_element = driver.find_element(By.CSS_SELECTOR, '#app > div > div.search-result-wrapper > div.search-list-wrapper > div > div.search-result-list > div.result-page > div > div > div.btn.page-next')
            new_element.click()
            sleep(1)

        
        except:
            print('kan niet vinden')
            break
            
        
scroll_page("JD.JD")
```
