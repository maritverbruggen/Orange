```python
def scroll_page(retailer):    
    
    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    num_products_view_2 = 0
    
    
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(200):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "cr-product-listing-wrapper w-100"})))
        print(num_products_view_1)
        
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                
                the_brand_id = soup.find_all(attrs={"class": "cr-product-listing-wrapper w-100"})[counter].find_all(attrs={"class": "cr-brand-name-wrapper"})[0].text
                the_brand_id = the_brand_id.replace("  ", "")
                
                the_name_id = soup.find_all(attrs={"class": "cr-product-listing-wrapper w-100"})[counter].find_all(attrs={"class": "product name"})[0].text
                the_name_id = the_name_id.replace("  ", "")
                the_name_id = the_brand_id + " " + the_name_id
                print(the_name_id)
                
                 
                the_price_id = soup.find_all(attrs={"class": "cr-product-listing-wrapper w-100"})[counter].find_all(attrs={"class": "price"})[0].text
                the_price_id = the_price_id.replace(" dt ", "")
                print(the_price_id)
                
                the_volume_id = soup.find_all(attrs={"class": "cr-product-listing-wrapper w-100"})[counter].find_all(attrs={"class": "cr-product-list-short-description-grid"})[0].text
                print(the_volume_id)
                
                filename = "Carrefour Tunesie"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per product into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
            
            try: 
                
                new_element = driver.find_element(By.CSS_SELECTOR, '#instant-search-pagination-container > div > ul > li.ais-Pagination-item.ais-Pagination-item--nextPage > a')
                driver.execute_script("arguments[0].click();", new_element)
                sleep(1)
            except: 
                print("Kan geen nieuwe pagina vinden")
                break 

    

scroll_page("Carrefour Tunesie")
```

```python
total_products = driver.find_element(By.CSS_SELECTOR, '#js-product-list-top > div.col-md-6.hidden-md-down.total-products').text
total_products = total_products.split()
total_products = total_products[0]
total_products = total_products.replace("(", "")
total_products = total_products.replace(")", "")
total_products = int(total_products)
print(total_products)
```

```python
def scroll_page(retailer):    
       
    sleep(1)
 
    
    num_products_view_2 = 0
    number = 0
    
    for _ in range(200):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of products in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "product-description"})))
        
        #create attributes the_name_id, the_price_id, the_brand_id and the_volume_id 
        for counter in range(num_products_view_2, num_products_view_1):
            the_brand_id = soup.find_all(attrs={"class": "product-description"})[counter].find_all(attrs={"class": "div_manufacturer_name"})[0].text
        
            
            the_name_id = soup.find_all(attrs={"class": "product-description"})[counter].find_all(attrs={"class": "h3 product-title"})[0].text
            the_name_id = the_name_id.replace("\n", "")
            the_name_id = the_name_id.replace("  ", "")
            the_name_id = the_brand_id + " " + the_name_id
            print(the_name_id)
            
            try:        
                the_price_id = soup.find_all(attrs={"class": "product-description"})[counter].find_all(attrs={"class": "regular-price"})[0].text
            except: 
                the_price_id = soup.find_all(attrs={"class": "product-description"})[counter].find_all(attrs={"class": "price"})[0].text
            print(the_price_id)
            
            
            the_volume = soup.find_all(attrs={"class": "product-description"})[counter].find_all(attrs={"class": "div_contenance"})[0].text
            the_volume = the_volume.replace("\n", "")
            the_volume = the_volume.replace("  ", "")
            print(the_volume)
            
            filename = "Monoprix Tunesie"
            fullpath = "../Output/" + str(filename) + ".csv"
            #write variables per product into csv file 
            with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                writer = csv.writer(csv_file)
                writer.writerow([retailer, the_brand_id, the_volume,the_name_id, the_price_id])
            number +=1
    
        try: 
            new_element = driver.find_element(By.CSS_SELECTOR, '#js-product-list > nav > div.col-md-8 > ul > li:nth-child(9) > a > i')
            new_element.click()
            sleep(2)
        except: 
            try: 
                new_element = driver.find_element(By.CSS_SELECTOR, '#js-product-list > nav > div.col-md-8 > ul > li:nth-child(8) > a')
                new_element.click()
                sleep(2)
            except: 
                try: 
                    new_element = driver.find_element(By.CSS_SELECTOR, '#js-product-list > nav > div.col-md-8 > ul > li:nth-child(7) > a > i')
                    new_element.click()
                    sleep(2)
                
                except: 
                    print("kan niet vinden")
        
        if number >= total_products:
            break 

        
scroll_page("Monoprix Tunesie")
```
