```python
def scroll_page(retailer):    
    
    #call get_url function to get the correct url linked to the Region that is going to be scraped    
    sleep(1)
    
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    scroll_range = 0
    num_products_view_2 = 0
    #create a scroller that scrolls the_range amount of times.     
    for _ in range(150):
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        
        # total number of wines in current view
        num_products_view_1 = int(len(soup.find_all(attrs={"class": "css-c3bgig"})))
        print(num_products_view_1)
        
        #create attributes the_name_id, the_price_id, the_reviews_id and the_rating id
        if num_products_view_1 > num_products_view_2:
            for counter in range(num_products_view_2, num_products_view_1):
                
                the_name_id = soup.find_all(attrs={"class": "css-c3bgig"})[counter].find_all(attrs={"class": "css-1nhiovu"})[0].find_all("a")[0].text
                print(the_name_id)
                try: 
                    the_price_id = soup.find_all(attrs={"class": "css-c3bgig"})[counter].find_all(attrs={"class": "css-iqeby6"})[0].text
                except: 
                    the_price_id = soup.find_all(attrs={"class": "css-c3bgig"})[counter].find_all(attrs={"class": "css-17fvam3"})[0].text
                
                the_brand_id = the_name_id.split()
                the_brand_id = the_brand_id[0] + " " + the_brand_id[1]
                
                the_volume_id = soup.find_all(attrs={"class": "css-c3bgig"})[counter].find_all(attrs={"class": "css-1ueix7b"})[0].text           
            
                filename = "Carrefour_Qatar"
                fullpath = "../Output/" + str(filename) + ".csv"
                #write variables per wine into csv file 
                with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
                    
            scroll_range += 300
            driver.execute_script('window.scrollTo(0, ' + str(scroll_range) + ')')

        #break als er geen "next" button gevonden kan worden  
        sleep(3)
        try:
            new_element = driver.find_element(By.CSS_SELECTOR, '#__next > div.css-qo9h12 > div > div.css-4jlmt > div.css-aa17wd > div.css-1hbp62g > button')
            new_element.click()
            sleep(1)

        except:
            print('kan niet vinden')
            scroll_range += 300
            driver.execute_script('window.scrollTo(0, ' + str(scroll_range) + ')')
            
        
        #update total number of products in current view to compare with number of products in view before scrolling
        num_products_view_2 = int(len(soup.find_all(attrs={"class": "css-u28fyu"})))
        print(num_products_view_2)

        #break loop if all products are scraped
        if num_products_view_2 >= total_products-20: 
            break
scroll_page("Carrefour_Qatar")
```
