```python
def scroll_page(retailer): 
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    
    num_products_view_2 = 0
    
    num_products_view_1 = int(len(soup.find_all(attrs={"class": "thumbnail-container"})))
        
    for counter in range(num_products_view_2, num_products_view_1):
        the_name_id = soup.find_all(attrs={"class": "thumbnail-container"})[counter].find_all(attrs={"class": "product-desc"})[0].text
        the_name_id = the_name_id.replace("\n", "")
        the_name_id = the_name_id.replace("  ", "")
        print(the_name_id)

        the_price_id = soup.find_all(attrs={"class": "thumbnail-container"})[counter].find_all(attrs={"class": "price"})[0].text
        the_price_id = the_price_id.replace("\n", "")
        the_price_id = the_price_id.replace("  ", "")
        print(the_price_id)
        the_brand_id = the_name_id
       
        the_volume_id = the_name_id 
        

        filename = "CIA Market Litouwen"
        fullpath = "../Output/" + str(filename) + ".csv"
        #write variables per product into csv file 
        with open(fullpath, mode='a', newline='', encoding='utf-8') as csv_file:
            writer = csv.writer(csv_file)
            writer.writerow([retailer, the_brand_id, the_volume_id,the_name_id, the_price_id])
            
scroll_page("CIA Market Litouwen")
```
